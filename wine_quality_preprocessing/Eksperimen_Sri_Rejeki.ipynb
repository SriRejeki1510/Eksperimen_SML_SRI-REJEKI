```python
# Eksperimen_Nama-siswa.ipynb

# --- 1. Persiapan: Import Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import os

print("Libraries imported successfully!")

# --- 2. Data Loading ---
# Mengunduh dan memuat dataset Wine Quality (Merah dan Putih)
# Sumber: https://archive.ics.uci.edu/dataset/186/wine+quality
red_wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
white_wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv'

try:
    df_red = pd.read_csv(red_wine_url, sep=';')
    df_red['type'] = 'red' # Tambahkan kolom 'type' untuk membedakan

    df_white = pd.read_csv(white_wine_url, sep=';')
    df_white['type'] = 'white' # Tambahkan kolom 'type'

    # Gabungkan kedua dataset
    df = pd.concat([df_red, df_white], ignore_index=True)
    print("Dataset Wine Quality (Red & White) berhasil dimuat!")
    print(f"Bentuk dataset gabungan: {df.shape}")
    print("\n5 baris pertama dataset gabungan:")
    print(df.head())
    print("\n5 baris terakhir dataset gabungan:")
    print(df.tail())

    # Simpan dataset raw ke folder namadataset_raw (khusus jika di VS Code)
    # Jika di Colab, Anda bisa lewati langkah ini untuk saat ini,
    # atau unduh manual dan letakkan di struktur folder lokal Anda.
    # Buat folder jika belum ada
    raw_data_dir = 'namadataset_raw'
    os.makedirs(raw_data_dir, exist_ok=True)
    df.to_csv(os.path.join(raw_data_dir, 'wine_quality_raw.csv'), index=False)
    print(f"\nDataset raw disimpan ke: {os.path.join(raw_data_dir, 'wine_quality_raw.csv')}")

except Exception as e:
    print(f"ERROR: Gagal memuat dataset. Pastikan URL benar dan koneksi internet tersedia. {e}")

# --- 3. Exploratory Data Analysis (EDA) ---
print("\n--- Exploratory Data Analysis (EDA) ---")

# Informasi Umum Dataset
print("\nInformasi Umum Dataset:")
df.info()

# Statistik Deskriptif Dataset
print("\nStatistik Deskriptif Dataset:")
print(df.describe())

# Pengecekan Nilai Hilang
print("\nJumlah Nilai Hilang per Kolom:")
print(df.isnull().sum())
# Dataset Wine Quality dikenal tidak memiliki nilai hilang, jadi outputnya harus 0 semua.

# Distribusi Variabel Numerik (Histogram)
print("\nVisualisasi Distribusi Variabel Numerik:")
df.hist(bins=15, figsize=(15, 10), layout=(3, 4))
plt.suptitle('Distribusi Fitur Numerik (Sebelum Preprocessing)', y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Adjust layout to prevent title overlap
plt.show()

# Distribusi Kolom 'quality'
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='quality', hue='type', palette='viridis')
plt.title('Distribusi Kualitas Anggur berdasarkan Tipe')
plt.xlabel('Kualitas (0-10)')
plt.ylabel('Jumlah')
plt.show()

# Korelasi Antar Fitur
print("\nMatriks Korelasi Antar Fitur Numerik:")
plt.figure(figsize=(12, 10))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriks Korelasi Antar Fitur')
plt.show()

# Distribusi 'type'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type', palette='pastel')
plt.title('Distribusi Tipe Anggur (Merah vs Putih)')
plt.show()

# Analisis Box Plot untuk Outlier (Beberapa Fitur Kunci)
print("\nAnalisis Box Plot untuk Mendeteksi Outlier:")
numeric_cols_for_boxplot = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
                            'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
                            'pH', 'sulphates', 'alcohol', 'quality']

for col in numeric_cols_for_boxplot:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='type', y=col, data=df, palette='Set2')
    plt.title(f'Box Plot of {col} by Wine Type')
    plt.show()

# --- 4. Data Preprocessing ---
print("\n--- Data Preprocessing ---")

# Definisikan fitur (X) dan target (y)
# 'quality' adalah target
# 'type' adalah fitur kategorikal yang perlu di-encode
# Fitur lainnya adalah numerik
X = df.drop('quality', axis=1)
y = df['quality']

# Identifikasi kolom numerik dan kategorikal
numerical_cols = X.select_dtypes(include=np.number).columns.tolist()
categorical_cols = ['type'] # Hanya 'type' karena sudah kita tambahkan

print(f"\nFitur Numerik: {numerical_cols}")
print(f"Fitur Kategorikal: {categorical_cols}")

# Buat preprocessor menggunakan ColumnTransformer
# Untuk fitur numerik: StandardScaler
# Untuk fitur kategorikal: OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

# Buat pipeline preprocessing
# Pipeline ini akan memastikan langkah-langkah preprocessing konsisten
preprocessing_pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Terapkan preprocessing pada seluruh dataset X untuk melihat hasilnya
# NOTE: Biasanya kita menerapkan fit_transform pada data training dan transform pada data testing.
# Untuk eksperimen ini, kita terapkan pada X keseluruhan untuk melihat bentuk akhir.
X_processed_array = preprocessing_pipeline.fit_transform(X)

# Dapatkan nama kolom setelah OneHotEncoding
encoded_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)
new_column_names = numerical_cols + list(encoded_feature_names)

X_processed_df = pd.DataFrame(X_processed_array, columns=new_column_names)

print("\nBentuk dataset X setelah preprocessing:")
print(X_processed_df.shape)
print("\n5 baris pertama dataset X setelah preprocessing:")
print(X_processed_df.head())
print("\nInformasi dataset X setelah preprocessing:")
X_processed_df.info() # Perhatikan bahwa semua kolom sekarang numerik dan tidak ada nilai hilang

# Pembagian Data: Training dan Testing
# Penting: Lakukan pembagian setelah preprocessing *fitur*, tetapi sebelum *transformasi target* jika target adalah kategorikal.
# Dalam kasus ini, target 'quality' adalah numerik (integer), jadi tidak perlu di-encode lebih lanjut.
X_train, X_test, y_train, y_test = train_test_split(X_processed_df, y, test_size=0.2, random_state=42)

print(f"\nBentuk X_train: {X_train.shape}")
print(f"Bentuk X_test: {X_test.shape}")
print(f"Bentuk y_train: {y_train.shape}")
print(f"Bentuk y_test: {y_test.shape}")

# Simpan data yang sudah diproses ke folder 'namadataset_preprocessing'
processed_data_dir = 'preprocessing/namadataset_preprocessing'
os.makedirs(processed_data_dir, exist_ok=True)

# Gabungkan X_processed_df dan y menjadi satu dataframe untuk penyimpanan
final_preprocessed_df = X_processed_df.copy()
final_preprocessed_df['quality'] = y.values # Pastikan indeks selaras

final_preprocessed_df.to_csv(os.path.join(processed_data_dir, 'wine_quality_preprocessed.csv'), index=False)
print(f"\nData preprocessed berhasil disimpan ke: {os.path.join(processed_data_dir, 'wine_quality_preprocessed.csv')}")

print("\nEksperimen di notebook selesai. Data siap untuk modelling!")
```
